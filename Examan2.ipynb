```python
import pandas as pd
from pandas_profiling import ProfileReport

import altair as alt
alt.renderers.enable('default')
```

```python
eventos_df = pd.read_csv('/data/EVENTOS_FEDERADOS_2022.csv')
```

```python
#Visualización #1: 
p1= alt.Chart(eventos_df).mark_bar().encode(
        alt.X('CATEGORIA'), 
        alt.Y('#'),
        tooltip = [
               alt.Tooltip('#')
        ]
    ).properties(
        width = 1000,
        height = 500
    )
p1.title ='Pregunta #1: ¿Cuántos eventos por categoría ha realizado o realizara la federación el presente año?'
p1
   
```

**Análisis e interpretación #1:** *En esta visualización podemos observar cuales son categorías sobre las que se ha realizado o realizaran los distintos eventos y de estas por la visualización podemos conocer el número de eventos programadas , para esto fue necesario tomar las columnas de categoría graficada en el eje X y # (que representa el número de eventos) en el eje Y*

**Pregunta #2:** *¿Cuál de las ciudades cubre la mayor parte de los eventos por la federación?*


```python
from typing import Text
from typing import Text
#Visualización #2: 
p2=alt.Chart(eventos_df).mark_bar().encode(    
    alt.X('count():Q'),                    
    alt.Y('CIUDAD:N', sort = '-x'), 
    alt.Color('CIUDAD:N')
).properties(
    width = 700,
    height = 550
).interactive()
p2.title ='Pregunta #2: ¿Cuál de las ciudades cubre la mayor parte de los eventos por la federación?'
p2
```

**Análisis e interpretación #2:**
*En esta visualización podemos observar  las distintas ciudades donde se han realizado o realizaran eventos en el país, mediante la visualización podemos observar de mayor a menor la ciudad y la frecuencia de los eventos , para esto fue necesario contar por ciudad cuántos eventos se han realizado como lo observamos anteriormente en el eje  X se grafica el conteo y el el Y la ciudad , adicional se le agrega color para identificar mas facil las mismas.*

Pregunta#3:¿Qué ligas se estarían jugando en la ciudad con más eventos y en qué fecha?

Filtro de los eventos en la ciudad de cali 

```python
cali_df = eventos_df[eventos_df['CIUDAD'] == 'CALI']
```

```python

p3=alt.Chart(cali_df).mark_bar().encode(
    x='LIGA:O',
    y='sum(#):Q',
    color='FECHA:N',
    column='CIUDAD:N'
).properties(
    width = 700,
    height = 150
).interactive()
p3.title ='Pregunta#3:¿Qué ligas se estarían jugando en la ciudad con más eventos y en qué fecha?'
p3
```

**Análisis e interpretación #3:** *Data la primera visualización la ciudad con más evento es cali, con esta información se filtra solo los eventos de la ciudad de cali para esta se analizan las ligas y las fechas que se estarían jugando cada una para esto fue necesario mostrar en la visualización esos datos configurados de la siguiente manera en el eje X la liga en el Y la suma de los eventos por liga , se clasifica por color las fechas y la agrupación de columna correspondiente a la ciudad de cali.*
